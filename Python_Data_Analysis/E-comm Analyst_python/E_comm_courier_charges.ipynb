# <font color='blue'><b>Calculate Weight for Slab</b></font>

```python
def weight_slab(weight):
  i=round(weight%1,1)
  print(i)
  if i==0.0:
    return weight
  elif i>0.5:
    return int(weight)+1.0
  else:
    return int(weight)+0.5
```

# <font color='red'>Claculated Slab New Column</font>

```python
SKU_pin_df['Weight Slab As per CC kg'] = SKU_pin_df['Total_Weight_per_shipment'].apply(weight_slab)
df_invoice['Weight slab charged by CC kg']  = df_invoice['Total Weight as per CC(kg)'].apply(weight_slab)
```

```python
df_CC_rates['Zone'] = df_CC_rates['Zone'].astype(str).str.strip().str.upper()
SKU_pin_df['Delivery_Zone_code'] = SKU_pin_df['Delivery_Zone_code'].astype(str).str.strip().str.upper()
```

```python
df_SKUPIN_mrg = df_CC_rates.merge(df_SKU_need_coll, on= "Delivery_Zone_code")
# print(df_SKUPIN_mrg.shape)
df_SKUPIN_mrg.head(5)
```

```python
SKU_pin_df.head(1)
```

```python
# Initialize an empty list to store the calculated expected charges
Total_Expected_Charge = []

# Loop through each row of the DataFrame
for i in range(len(df_CC_rates)):
    zone = df_SKUPIN_mrg['Delivery_Zone_code'][i]
    weight_slabs = df_SKUPIN_mrg['Weight Slab As per CC kg'][i]
    # print(weight_slabs)
    fwd_fixed = df_CC_rates['Forward Fixed Charge'][i]
    # print(fwd_fixed)
    fwd_additional = df_CC_rates['Forward Additional Weight Slab Charge'][i]
    rto_fixed = df_CC_rates['RTO Fixed Charge'][i]
    rto_additional = df_CC_rates['RTO Additional Weight Slab Charge'][i]
    shipment_type = SKU_pin_df['Type of Shipment'][i]

    if shipment_type == 'Forward charges':
        if weight_slabs <= 0.5:
            Total_Expected_Charge.append(fwd_fixed)
            # print(Total_Expected_Charge)
        else:
            additional_weight = (weight_slabs - 0.5) / 0.5
            Total_Expected_Charge.append(fwd_fixed + additional_weight * fwd_additional)
            # print(Total_Expected_Charge)
    elif shipment_type == 'Forward and RTO charges':
        if weight_slabs <= 0.5:
            Total_Expected_Charge.append(fwd_fixed + rto_fixed)
        else:
            additional_weight = (weight_slabs - 0.5) / 0.5
            Total_Expected_Charge.append(fwd_fixed + additional_weight * (fwd_additional + rto_additional))

# Update the 'Expected Charge As Per ABC (Rs.)' column in the DataFrame
# print(Total_Expected_Charge)
df_CC_rates['Expected Charge as per X (Rs.)'] = Total_Expected_Charge
```

# <font color='green'><b>*Difference Between Expected Charges And Billed Charges*</b></font>

```python
Output_res_1['Difference Between Expected Charges and Billed Charges (Rs.)'] = ''
Output_res_1['Difference Between Expected Charges and Billed Charges (Rs.)'] = (Output_res_1['Charges Billed by Courier Company (Rs.)'] - Output_res_1['Expected Charge as per X (Rs.)'])
Output_res_1['Difference Between Expected Charges and Billed Charges (Rs.)']
```

```python
Output_result_1 = Output_res_1[['Order ID','AWB_Number','Total_Weight_per_shipment','Weight Slab As per CC kg','Total Weight as per CC(kg)','Weight slab charged by CC kg','Delivery_Zone_code','Delivery zone Chargerd by CC(Kg)','Type of Shipment','Expected Charge as per X (Rs.)','Charges Billed by Courier Company (Rs.)','Difference Between Expected Charges and Billed Charges (Rs.)']]
Output_result_1.head(1)
```

```python
correctly_charged=Output_res_1[Output_res_1['Difference Between Expected Charges and Billed Charges (Rs.)']==0.0]
correctly=["Total Orders - Correctly Charged",len(correctly_charged),sum(correctly_charged['Charges Billed by Courier Company (Rs.)'])]

over_charged=Output_res_1[Output_res_1['Difference Between Expected Charges and Billed Charges (Rs.)']<0.0]
over=["Total Orders - Over Charged",len(over_charged),np.absolute(sum(over_charged['Difference Between Expected Charges and Billed Charges (Rs.)']))]

under_charged=Output_res_1[Output_res_1['Difference Between Expected Charges and Billed Charges (Rs.)']>0.0]
under=["Total Orders - Under Charged",len(under_charged),np.absolute(sum(under_charged['Difference Between Expected Charges and Billed Charges (Rs.)']))]
```

```python
Output_res_2 = pd.DataFrame([correctly,over,under], columns= ['Discription','Count','Amount (Rs.)'])
```
