```python
pwd = os.getcwd()                               # current working directory
print(pwd)
mobile_df = pwd + "/Flipkart_Mobiles.csv"          # assigning file to respective variable
```

```python
data = pd.read_csv(mobile_df,delimiter=',')
data.head()
```

```python
sns.pairplot(data,hue='Brand',size=None)
```

```python
sns.displot(data,x='Selling Price',bins=[5000,10000,15000,20000,25000,30000,35000,40000,50000,60000,80000],color='orange',aspect=1.7)
```

```python
sns.displot(data,x='Original Price',bins=[5000,10000,15000,20000,25000,30000,35000,40000,50000,60000,80000],color='#59ff17',aspect=1.7)
```

```python
sns.displot(data,x='Selling Price', bins=5,hue='Brand',aspect=1.2)    #brandwise mobile phone price segments
```

```python
fig, ax = plt.subplots(figsize=(15,3))
ax=sns.countplot(x="Brand", data=data)
```

```python
fig, ax = plt.subplots(figsize=(15,3))
ax = sns.stripplot(y="Rating", x="Brand", data=data)
```

```python
fig, ax = plt.subplots(figsize=(15,3))
ax = sns.stripplot(y="Selling Price", x="Brand", data=data)
```

```python
fig, ax = plt.subplots(figsize=(15,3))
ax = sns.stripplot(y="Original Price", x="Brand", data=data)
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax = sns.boxplot(x="Brand", y="Selling Price", data=data)
```

```python
round(data.groupby('Brand')['Selling Price'].mean(),0).sort_values(ascending=True)   # Brandwise Mean Selling prices
```

```python
round(data.groupby('Brand')['Original Price'].mean(),0).sort_values(ascending=True)   # Brandwise Mean Selling prices
```
